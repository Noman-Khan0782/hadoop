```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets (Replace 'states_all.csv' and 'US_violent_crime.csv' with your actual file paths)
data_1 = pd.read_csv('states_all.csv')
data_2 = pd.read_csv('US_violent_crime.csv')
```

```python
# Rename columns for consistency
data_2.rename(columns={'Unnamed: 0': 'STATE'}, inplace=True)
```

```python
# Rename columns for consistency
data_2.rename(columns={'Unnamed: 0': 'STATE'}, inplace=True)
```

```python
# Standardize state names in both datasets
data_1['STATE'] = data_1['STATE'].str.upper()
data_2['STATE'] = data_2['STATE'].str.upper()
```

```python
# Merge datasets on the 'STATE' column
merged_data = pd.merge(data_1, data_2, on='STATE',how='inner')
```

```python
merged_data = pd.merge(data_1, data_2, on='STATE', how='inner')
```

```python
# Handle missing data: Fill missing numerical data with the median
numerical_columns = merged_data.select_dtypes(include=['float64', 'int64']).columns
merged_data[numerical_columns] = merged_data[numerical_columns].fillna(merged_data[numerical_columns].median())
```

```python
# 1. Descriptive Statistics
print("Summary Statistics:")
print(merged_data.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 2. Yearly Trends for Violent Crimes
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged_data, x='YEAR', y='Murder', label='Murder')
sns.lineplot(data=merged_data, x='YEAR', y='Assault', label='Assault')
sns.lineplot(data=merged_data, x='YEAR', y='Rape', label='Rape')
plt.title('Yearly Trends in Violent Crimes')
plt.xlabel('Year')
plt.ylabel('Crime Rates')
plt.legend()
plt.show()
```
